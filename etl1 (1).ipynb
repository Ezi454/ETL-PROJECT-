```python
import pandas as pd
import psycopg2
from psycopg2.extras import execute_values
```

```python
df = pd.read_csv(r"C:\Users\ikhra\Downloads\penjualan_100.csv")
```

```python
df.isnull().sum()
```

```python
df["Total_Penjualan"] = df["jumlah_terjual"] * df["harga_satuan"]
```

```python
df.drop(columns=["Harga Baru"], inplace=True)
```

```python
df
```

```python
con = psycopg2.connect(
    host = 'localhost',
    database = 'project_1',
    user = 'ikhram_user',
    password = 'password123')
```

```python
cur = con.cursor()
```

```python
data_tuples = list(df.itertuples(index=False, name=None))
```

```python
insert_query = """
INSERT INTO data_penjualan (order_id, tanggal_order, produk, kategori, jumlah_terjual, harga_satuan, Total_Penjualan)
VALUES %s
ON CONFLICT (order_id) DO NOTHING;
"""
```

```python
execute_values(cur, insert_query, data_tuples)
con.commit()
```

```python
cur.close()
con.close()

print("Data berhasil diimport ke PostgreSQL!")
```
